# Guess Game Program

```python
correct_num = 9

guess_num = 0

while guess_num != correct_num:
    guess_num = int(input("Guess the number: "))

    print("Wrong, try again" if guess_num != correct_num else "You guessed it")

print("Thank you for playing")
```
